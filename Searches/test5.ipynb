```python
import numpy as np
import matplotlib.pyplot as plt
spaced=np.linspace(-10,10,50)
sine=np.sin(spaced)
plt.plot(spaced,sine)
plt.xlabel("Angle(Radians)")
plt.ylabel("Sine value")
plt.title("Line chart")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
a=np.array([1,3,5,7,9])
b=np.array([2,4,6,8,10])
Add=np.add(a,b)
Sub=np.subtract(a,b)
Mul=np.multiply(a,b)
Div=np.divide(a,b)
plt.plot(a, Add, label="Addition")
plt.plot(a, Sub, label="Subtraction")
plt.plot(a, Mul, label="Multiplication")
plt.plot(a, Div, label="Division")
plt.xlabel("Operations")
plt.ylabel("Array elements")
plt.title("Mathematical Operations")
plt.legend()
plt.show()
```

```python
arr=np.linspace(0,2*np.pi,100)
sin=np.sin(arr)
cos=np.cos(arr)
print(sin)
print(cos)
plt.plot(arr,sin,label="Sine",color="green")
plt.plot(arr,cos,label="cos",color="blue")
plt.xlabel("Angles")
plt.ylabel("Values")
plt.title("Sin and cos values")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
arr=np.arange(1,11)
square=arr**2
print(square)
plt.plot(arr,square,linestyle="--",marker="o",label="n**2")
plt.xlabel("input")
plt.ylabel("Squared Value")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
arr1=np.random.randint(1,50,20)
arr2=np.random.randint(10,100,20)
sort_arr1=np.sort(arr1)
sort_arr2=np.sort(arr2)
print(sort_arr1)
print(sort_arr2)
plt.plot(arr1,linestyle="-",label="1-50",marker="o")
plt.plot(arr2,linestyle="--",label="10-100",marker="v")
plt.xlabel("index")
plt.ylabel("Values")
plt.title("Comparing two trends")
plt.legend()
plt.show()
```
