```python
Problem Statement: Write a program that takes a string as input and returns a dictionary where the keys are the characters in the string, 
and the values are the number of times each character appears.
Ignore spaces and make the counting case-insensitive.
Example Input: "Programming is fun"
Example Output: {'p': 1, 'r': 2, 'o': 1, 'g': 2, 'a': 1, 'm': 2, 'i': 2, 'n': 2, 's': 1, 'f': 1, 'u': 1}
```

```python
s="Programming is Programming"
def char_repeat(s):
    char_count = {}
    for char in s:
        char_count[char] = char_count.get(char, 0) + 1
    return char_count
result = char_repeat(s)
print(result)
```

```python
Problem Statement: Write a program that takes a list of strings and groups them into dictionaries based on anagrams.
The dictionary keys should be the sorted version of the anagrams, and the values should be lists of words that are anagrams of each other.
Example Input: ["listen", "silent", "enlist", "rat", "tar", "art", "hello"]
Example Output: {'eilnst': ['listen', 'silent', 'enlist'], 'art': ['rat', 'tar', 'art'], 'ehllo': ['hello']}
```

```python
x=["listen", "silent", "enlist", "rat", "tar", "art", "hello"]

def anagrams(x):
    y= {}
    for word in x:
        sort = ''.join(sorted(word))
        if sort not in y:
            y[sort] = []
        y[sort].append(word)
    return y
result = anagrams(x)
print(result)



```

```python
Problem Statement: Write a program that takes a string and a list of substrings as input and returns a dictionary
where the keys are the substrings and the values are the number of times each substring appears in the string.
Example Input:
String: "banana bandana"
Substrings: ["ban", "ana", "na"]
Example Output: {'ban': 2, 'ana': 2, 'na': 3}
```

```python
s1 = "banana bandana"
substrings = ["ban", "ana", "na"]

result = {}
for substring in substrings:
    result[substring] = s1.count(substring)

print(result)
```
